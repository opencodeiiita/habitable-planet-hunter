```python
    #filling missing values with mean
df['P_MASS'] = df['P_MASS'].fillna(df['P_MASS'].mean())
```

```python
df['P_MASS'].isnull().sum()
```

```python
df['P_MASS_ERROR_MIN'].mean()
```

```python
df['P_MASS_ERROR_MIN'] = df['P_MASS_ERROR_MIN'].fillna(df['P_MASS_ERROR_MIN'].mean())
df['P_MASS_ERROR_MIN'].isnull().sum()
```

```python
df['P_MASS_ERROR_MAX'] = df['P_MASS_ERROR_MIN'].fillna(df['P_MASS_ERROR_MIN'].median())
```

```python
#dropping irrelevant columns
df=df.drop(['P_SEMI_MAJOR_AXIS_EST'],axis=1)
```

```python
#univariate analysis
plt.figure(figsize=(10,6))
sns.histplot(df['P_MASS'], bins=30, kde=True)
plt.title('Distribution of Planet Mass')
plt.xlabel('Planet Mass (Earth Masses)')
```

```python
#univariate analysis for all
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
for col in numerical_cols:
    plt.figure(figsize=(10,6))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.show()
    #categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:    
    plt.figure(figsize=(10,6))
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
#bivariate analysis
plt.figure(figsize=(10,6))
sns.scatterplot(x='P_TEMP_EQUIL', y='P_MASS', data=df)
plt.title('Planet Mass vs Equilibrium Temperature')
plt.xlabel('Equilibrium Temperature (K)')
plt.ylabel('Planet Mass (Earth Masses)')
plt.show()
```

```python
#checking outliers
plt.figure(figsize=(10,6))
sns.boxplot(x=df['P_MASS'])
plt.title('Boxplot of Planet Mass')
plt.xlabel('Planet Mass (Earth Masses)')
plt.show()
```

```python
#removing outliers
Q1 = df['P_MASS'].quantile(0.25)
Q3 = df['P_MASS'].quantile(0.75)    
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['P_MASS'] >= lower_bound) & (df['P_MASS'] <= upper_bound)]
```

```python
#Testing after removing outliers
plt.figure(figsize=(10,6))
sns.boxplot(x=df['P_MASS'])
plt.title('Boxplot of Planet Mass After Removing Outliers')
plt.xlabel('Planet Mass (Earth Masses)')
plt.show()
```

```python
#target variable analysis
plt.figure(figsize=(10,6))
sns.histplot(df['P_MASS'], bins=30, kde=True)
plt.title('Distribution of Planet Mass After Outlier Removal')
plt.xlabel('Planet Mass (Earth Masses)')
plt.show()
```
